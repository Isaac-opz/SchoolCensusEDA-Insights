```python
import pandas as pd

# Carga el archivo CSV
df = pd.read_csv('data/sample3.1.csv', encoding='ISO-8859-1')

# Reemplaza los espacios vacíos con NaN
df.replace('', pd.NA, inplace=True)

# Elimina las columnas con más de un 50% de datos faltantes
df.dropna(axis=1, thresh=df.shape[0]*0.5, inplace=True)

# Elimina las filas con más de un 50% de datos faltantes
df.dropna(axis=0, thresh=df.shape[1]*0.5, inplace=True)


df.dropna(inplace=True)
# Tratamiento de valores atípicos para columnas numéricas
for col in df.select_dtypes(include=['float64', 'int64']).columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Opción 1: Elimina valores atípicos
    df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]


# Guarda el dataframe limpio a un nuevo archivo CSV
df.to_csv('data/sample3.1_cleaned.csv', index=False)

df = pd.read_csv('data/sample3.1_cleaned.csv', encoding='ISO-8859-1')

df
```

```python
import pandas as pd

# Carga el archivo CSV
df = pd.read_csv('data/sample3.2.csv', encoding='ISO-8859-1')

# Reemplaza los espacios vacíos con NaN
df.replace('', pd.NA, inplace=True)

# Elimina las columnas con más de un 50% de datos faltantes
df.dropna(axis=1, thresh=df.shape[0]*0.5, inplace=True)

# Elimina las filas con más de un 50% de datos faltantes
df.dropna(axis=0, thresh=df.shape[1]*0.5, inplace=True)


df.dropna(inplace=True)
# Tratamiento de valores atípicos para columnas numéricas
for col in df.select_dtypes(include=['float64', 'int64']).columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Opción 1: Elimina valores atípicos
    df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]


# Guarda el dataframe limpio a un nuevo archivo CSV
df.to_csv('data/sample3.2_cleaned.csv', index=False)

df = pd.read_csv('data/sample3.2_cleaned.csv', encoding='ISO-8859-1')

df
```

We have 412 rows of data in total, so we can now concatenate the cleaned data in one single sample with 300 rows of data

```python
import pandas as pd

# Carga los archivos CSV
df1 = pd.read_csv('data/sample3.1_cleaned.csv', encoding='ISO-8859-1')
df2 = pd.read_csv('data/sample3.2_cleaned.csv', encoding='ISO-8859-1')


# Concatena los DataFrames
df_concatenado = pd.concat([df1, df2], ignore_index=True)

# Limita el DataFrame concatenado a las primeras 300 filas
df_limitado = df_concatenado.head(300)

# Guarda el DataFrame limitado en un nuevo archivo CSV
df_limitado.to_csv('data/sample3.csv', index=False)

# Imprime la información del DataFrame limitado
#print(f"El archivo limitado tiene {df_limitado.shape[0]} filas y {df_limitado.shape[1]} columnas.")
df = pd.read_csv('data/sample3.csv', encoding='ISO-8859-1')
df
```
